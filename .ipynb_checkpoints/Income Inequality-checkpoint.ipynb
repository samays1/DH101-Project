# Research Question:
Income inequality is less talked about/visualized than wealth inequality. The difference is that wealth is a measure of net worth, where as income is the amount of money that you earn in any given year. Because the super rich can pass down so much wealth, and the rich tend to get richer, wealth inequality is generally considered a bigger issue. Additionally, statistics cited on wealth and income inequality generally discuss the top 1% to 5% of the population. Both of these factors lead to a bit of missing information on how income inequality has affected those in the middle of the population, say the 10th to 90th percentile of earners. 

Because of this lack of information I have decided to ask: how has income inequality changed for the middle of the population over the last 25 years? How were the changes different under different administrations?

## Data: 
All data for this project was pulled from The Bureau of Labor's website. It contains annual income information for 100s of different job titles in the U.S. over the last 22 years.

# Methods:
I will clean all of the data, and track how income has changed for those that fall into the 10th, 25th, 50th, 75th, and 90th percentiles within their respective careers. 

```python
import pandas as pd
import numpy as np
from io import StringIO
import matplotlib.pyplot as plt
```

```python
list_of_dfs = []
```

Reading in each dataset (1 per year) for an excel file, and doing some cleaning so that each table will have the same column names. Each dataframe is saved into the list that is instantiated above:

```python
for i in np.arange(1997, 2014):
    new_df = pd.read_excel('DIGHUM data/Excel data/national_' + str(i) + '_dl.xls')
    new_df.columns = [x.lower() for x in new_df.columns]
    columns = new_df.columns
    if 'year' in columns:
        new_df.drop(columns = 'year')
    new_col = []
    for j in np.arange(len(columns)):
        temp = columns[j].replace('w','')
        new_col.append(temp)
    new_df.columns = new_col
    year_col = np.empty(len(new_df))
    for j in range(len(new_df)):
        year_col[j] = i
    new_df['year'] = year_col
    list_of_dfs.append(new_df)
```

Reading in data from years with an xlsx file:

```python
for i in np.arange(2014, 2019):
    new_df = pd.read_excel('DIGHUM data/Excel data/national_' + str(i) + '_dl.xlsx')
    new_df.columns = [x.lower() for x in new_df.columns]
    columns = new_df.columns
    if 'year' in columns:
        new_df.drop(columns = 'year')
    new_col = []
    for j in np.arange(len(columns)):
        temp = columns[j].replace('w','')
        new_col.append(temp)
    new_df.columns = new_col
    year_col = np.empty(len(new_df))
    for j in range(len(new_df)):
        year_col[j] = i
    new_df['year'] = year_col
    list_of_dfs.append(new_df)
```

Making sure all of the data frames were read in:

```python
print(len(list_of_dfs))
```

Concatenating the dataframes into one large df:

```python
big_table = pd.concat(list_of_dfs, sort = False);
big_table.head()
```

Selecting the columns that contain relevant data:

```python
annual_table = big_table[['occ_title', 'tot_emp', 'a_mean', 'a_pct10', 'a_pct25','a_median', 'a_pct75', 'a_pct90', 'year']]
```

Analyzing the df's dimensions:

```python
annual_table.shape
```

Removing rows that have non-numeric values for percentiles:

```python
for string in ['a_pct10', 'a_pct25', 'a_median', 'a_pct75', 'a_pct90', 'tot_emp']:
    annual_table = annual_table[pd.to_numeric(annual_table[string], errors='coerce').notnull()]
    list_version = annual_table[string].tolist()
    allInts = list(map(int, list_version))
    annual_table[string] = allInts
```

Removing rows that have NAN in any column:

```python
annual_table = annual_table.dropna(axis = 'rows', how = 'any')
```

Making all job titles lower case, removing duplicate rows, and resetting the index:

```python
title_list = annual_table['occ_title'].tolist()
title_list = [x.lower() for x in title_list]
annual_table['occ_title'] = title_list
annual_table = annual_table.drop_duplicates();
annual_table = annual_table.reset_index(drop= True);
```

Let's see what the table looks like after all that work!

```python
annual_table.reset_index(drop=True)
```

Lets take a peak at which jobs have the highest pay median income:

```python
annual_table.sort_values(by = 'a_median', ascending = False)
```

Before we track how income inequality has changed, we need to make sure that job titles aren't changing from year to year let's see how many job titles there are in each year as of right now:

```python
counted = annual_table.groupby('year').count()
counted = counted[['occ_title']]
counted.rename({'occ_title': 'number of titles'}, axis = 'index')
```

Now let's only keep titles that appear in all of the years:

```python
def getYearData(year):
    years = annual_table.loc[annual_table['year'] == year]
    return set(years['occ_title'].tolist())

titlesInYear = []
for year in counted.index.values.tolist():
    titles = getYearData(year)
    titlesInYear.append(titles)

commonSet = titlesInYear[0]
for i in range(len(titlesInYear)-1):
    commonSet = commonSet.intersection(titlesInYear[i])

consistentDF = annual_table[annual_table['occ_title'].isin(list(commonSet))]
consistentDF = consistentDF.reset_index(drop = True)
```

Let's see how many titles remain:

```python
len(commonSet)
```

Writing a function to find the overall average median, 10th, 25th etc. percentile in any given year (weighted by the number of people in the occupation)

```python
def findAverage(year, percentile):
    indexed_by_year = consistentDF.loc[consistentDF['year'] == year]
    income = indexed_by_year[percentile].values
    numEmployed = indexed_by_year['tot_emp'].values
    totalIncome = []
    for i in range(0, len(income)):
        totalIncome.append(income[i] * numEmployed[i])
    return sum(totalIncome) / sum(numEmployed)
```

Creating lists with the average 10th, 25th etc. percentiles for each year:

```python
tenth = []
twenthyfifth = []
median = []
seventyfifth = []
ninetyith = []
colNames = ['a_pct10', 'a_pct25', 'a_median', 'a_pct75', 'a_pct90']
for i in np.arange(1997.0, 2000.0):
    tenth.append(findAverage(i, colNames[0]))
for i in np.arange(2001.0, 2002.0):
    tenth.append(findAverage(i, colNames[0]))
for i in np.arange(2003.0, 2019.0):
    tenth.append(findAverage(i, colNames[0]))
    
for i in np.arange(1997.0, 2000.0):
    twenthyfifth.append(findAverage(i, colNames[1]))
for i in np.arange(2001.0, 2002.0):
    twenthyfifth.append(findAverage(i, colNames[1]))
for i in np.arange(2003.0, 2019.0):
    twenthyfifth.append(findAverage(i, colNames[1]))
    
for i in np.arange(1997.0, 2000.0):
    median.append(findAverage(i, colNames[2]))
for i in np.arange(2001.0, 2002.0):
    median.append(findAverage(i, colNames[2]))
for i in np.arange(2003.0, 2019.0):
    median.append(findAverage(i, colNames[2]))
    
for i in np.arange(1997.0, 2000.0):
    seventyfifth.append(findAverage(i, colNames[3]))
for i in np.arange(2001.0, 2002.0):
    seventyfifth.append(findAverage(i, colNames[3]))
for i in np.arange(2003.0, 2019.0):
    seventyfifth.append(findAverage(i, colNames[3]))
    
for i in np.arange(1997.0, 2000.0):
    ninetyith.append(findAverage(i, colNames[4]))
for i in np.arange(2001.0, 2002.0):
    ninetyith.append(findAverage(i, colNames[4]))
for i in np.arange(2003.0, 2019.0):
    ninetyith.append(findAverage(i, colNames[4]))
```

Making a list of all the years that have data remaining and then plotting:

```python
years = np.arange(1997, 2019).tolist()
years.remove(2000)
years.remove(2002)
```

```python
plt.figure(figsize=(9, 6))
plt.plot(years, ninetyith, 'g', years, seventyfifth, 'r', years, median, 'b', years, twenthyfifth, 'cyan', years, tenth, 'y')
plt.gca().legend(('90th percentile','75th percentile', '50th percentile', '25th percentile', '10th percentile'))
plt.title('Annual Income in the US')
plt.xlabel('Year')
plt.ylabel('Income (USD)')
plt.ticklabel_format(useOffset = False);
```

Finding the difference between the average 10th and 90th percentile income for each year and plotting:

```python
diff = []
for i in np.arange(len(ninetyith)):
    diff.append(ninetyith[i] - tenth[i])
```

```python
plt.figure(figsize=(9, 6))
plt.plot(years, diff)
plt.title('Difference between 10th and 90th percentile income in US')
plt.xlabel('year')
plt.ylabel('Difference (USD)')
plt.ticklabel_format(useOffset = False);
```

Wow, the inequality among average Americans does seem to be growing massively as well. Here is how much exactly:

```python
diff[19] / diff[0]
```

Now finding the average change in the difference between the 10th and 90th percentile per year for each of the last four administrations. (important to note the relatively small sample sized for Clinton and Trump

```python
clinton = (diff[2] - diff[0]) / 3
bush = (diff[9] - diff[3]) / 7
obama = (diff[17] - diff[10]) / 8
trump = (diff[19] - diff[18]) / 2
```

```python
objects = ('Clinton', 'Bush', 'Obama', 'Trump')
y_pos = np.arange(len(objects))
performance = [clinton, bush, obama, trump]

plt.bar(y_pos, performance, align='center', alpha=0.5, color=['black', 'red', 'green', 'blue'])
plt.xticks(y_pos, objects)
plt.ylabel('Avg Annual Increase (USD)')
plt.title('Average Yearly Increase in Difference Between 10th and 90th Percentile Income')

plt.show()
```

# Conclusion:
It is unfortunate that so much data was lost in the process of ensuring its accuracy. 180 job titles are not enough to make sweeping generalizations about America as a whole, but it is enough give us an interesting peek into the overall trend that income has had in the last 22 years.

The difference between the 10th and 90th percentile income has grown significantly more than I had anticipated, increasing by over 50% from 1997 to 2018. The inequality increased the most per year under Bush and Clinton, but Clinton and Trump have small sample sizes, and there are so many factors at play that it we cannot make serious claims as to how the governing style and economic practices of different administrations affects income inequality. 

There does seem to be an outlier year in the data set, as 2012 saw a dip in wages for every percentile, and the inequality as a whole. It is hard to say for certain what the cause is, but it could have to do with the election season, though that is not reflected in any of the other election years. It is also possible that the discrepecny in the data is a result of mismeasurement or some other source of unreliablity within the data set, so it is important that we remain skeptical about the data that we use, and avoid making serious claims without doing due diligence to ensure that they are founded.
